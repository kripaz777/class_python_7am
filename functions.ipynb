```python
def cal(y):
    l,b,h = y
#     l = y[0]
#     b = y[1]
#     h = y[2]
    a = l*b
    v = a*h
    return [a,b]

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
x = (l,b,h)
area,volume = cal(x)
print("The area is = ",area)
print("The volume is = ",volume)
```

```python
def cal(y):
    l,b,h = y
#     l = y[0]
#     b = y[1]
#     h = y[2]
    a = l*b
    v = a*h
    return {"area":a,"volume":v}

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
x = (l,b,h)
k = cal(x)
print("The area is = ",k['area'])
print("The volume is = ",k['volume'])
```

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b
    
def div(a,b):
    return a/b

def err():
    print("value of b can not be zero.")

def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * / ")
    if o =='+':
        print(add(a,b))
    elif o =='-':
        print(sub(a,b))
    elif o =='*':
        print(mul(a,b))
    elif o =='/':
        if b!=0:
            print(div(a,b))
        else:
            err()
    else:
        print("Invalid Operator")
cal()
```
